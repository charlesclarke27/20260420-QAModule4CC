```python
import pandas as pd
import numpy as np
import re
```

```python
url = "https://datasets.imdbws.com/title.basics.tsv.gz"

df = pd.read_csv(
    url,
    sep="\t",
    compression="gzip",
    low_memory=False
)

df.head()
```

```python
print(df.shape)
print(df.info())
print(df.isnull().sum())
```

```python
# Drop rows missing critical info
df = df.dropna(subset=["primaryTitle", "startYear"])

# Fill non-critical missing values
df["genres"] = df["genres"].fillna("Unknown")
df["runtimeMinutes"] = df["runtimeMinutes"].replace("\\N", np.nan)
```

```python
df = df.drop_duplicates()
```

```python
df["titleType"] = df["titleType"].str.lower().str.strip()
df["genres"] = df["genres"].str.lower().str.strip()
```

```python
df["startYear"] = pd.to_numeric(df["startYear"], errors="coerce")
df["runtimeMinutes"] = pd.to_numeric(df["runtimeMinutes"], errors="coerce")
df["isAdult"] = pd.to_numeric(df["isAdult"], errors="coerce").fillna(0).astype(int)
```
